```python
def residual_tcn_block(x, filters, kernel_size, dilation, dropout=0.1):

    res = x

    # 1er conv causal dilaté
    x = layers.Conv1D(
        filters,
        kernel_size,
        padding="causal",
        dilation_rate=dilation
    )(x)
    x = layers.ReLU()(x)
    x = layers.Dropout(dropout)(x)

    # 2e conv causal dilaté
    x = layers.Conv1D(
        filters,
        kernel_size,
        padding="causal",
        dilation_rate=dilation
    )(x)
    x = layers.ReLU()(x)
    x = layers.Dropout(dropout)(x)

    # projection si mismatch dimensions
    if res.shape[-1] != filters:
        res = layers.Conv1D(filters, 1, padding="same")(res)

    # residual connection
    x = layers.Add()([x, res])
    x = layers.ReLU()(x)

    return x
```

```python
def build_tcn(
    seq_len,
    n_past_features,
    horizon,
    n_future_features,
    channels=(32, 64),
    kernel_size=3,
    dropout=0.1
):

    # INPUT PASSÉ
    past_input = layers.Input(shape=(seq_len, n_past_features))
    x = past_input

    for i, ch in enumerate(channels):
        x = residual_tcn_block(
            x,
            filters=ch,
            kernel_size=kernel_size,
            dilation=2**i,
            dropout=dropout
        )

    x = layers.Lambda(lambda t: t[:, -1, :])(x)


    # INPUT FUTUR (covariables connues)
    future_input = layers.Input(shape=(horizon, n_future_features))

    f = layers.Flatten()(future_input)
    f = layers.Dense(32, activation="relu")(f)


    # FUSION
    h = layers.Concatenate()([x, f])

    outputs = layers.Dense(horizon)(h)

    return Model(inputs=[past_input, future_input], outputs=outputs)
```

```python
def build_tcn_multi_horizon(
    seq_len,
    n_features,
    horizon=10,
    channels=(32, 64),
    kernel_size=3,
    dropout=0.1
):

    inputs = layers.Input(shape=(seq_len, n_features))
    x = inputs

    for i, ch in enumerate(channels):
        x = residual_tcn_block(
            x,
            filters=ch,
            kernel_size=kernel_size,
            dilation=2**i,
            dropout=dropout
        )

    x = layers.Lambda(lambda t: t[:, -1, :])(x)

    outputs = layers.Dense(horizon)(x)  # multi-step direct

    return Model(inputs, outputs)
```

```python
def predict_tcn(model, X_last, n_steps, future_cov=None):
    """
    Prediction recursive pour TCN
    
    Parameters
    ----------
    model : keras model
    X_last : dernière fenêtre (1, seq_len, n_features)
    n_steps : nombre de pas à prédire
    future_cov : covariables futures (n_steps, n_cov) ou None
    strategy:
        - "recursive" -> 1-step model
        - "direct" -> multi-horizon model
    
    Returns
    -------
    preds : (n_steps,1)
    """

    if strategy == "direct":
        return model(X_last, training=False).numpy()

    elif strategy == "recursive":
        preds = []
        X_window = X_last.copy()

        for _ in range(n_steps):
            y = model(X_window, training=False).numpy()[0, 0]
            preds.append(y)

            X_window = np.roll(X_window, -1, axis=1)
            X_window[0, -1, 0] = y

        return np.array(preds).reshape(-1, 1)

    else:
        raise ValueError("strategy must be 'recursive' or 'direct'")
```

```python
def fit_model(model, X_train, y_train, X_val=None, y_val=None,
              epochs=50, batch_size=32, lr=1e-3):

    model.compile(
        optimizer=tf.keras.optimizers.Adam(lr),
        loss="mse"
    )

    model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val) if X_val is not None else None,
        epochs=epochs,
        batch_size=batch_size,
        shuffle=False
    )

    return model
```

```python
def create_sequences(X, y, seq_len):
    Xs, ys = [], []

    X = X.values if hasattr(X, "values") else X
    y = y.values if hasattr(y, "values") else y

    for i in range(len(X) - seq_len):
        Xs.append(X[i:i+seq_len])
        ys.append(y[i+seq_len])

    return np.array(Xs), np.array(ys).reshape(-1, 1)
```

```python
def make_dataset(data, seq_len, horizon):

    values = data.values  # (time, features)

    X_past, X_future, y_out = [], [], []

    for i in range(len(values) - seq_len - horizon):

        past = values[i : i + seq_len, 1:]  # covariables
        future = values[i + seq_len : i + seq_len + horizon, 1:]
        target = values[i + seq_len : i + seq_len + horizon, 0]

        X_past.append(past)
        X_future.append(future)
        y_out.append(target)

    return (
        np.array(X_past),
        np.array(X_future),
        np.array(y_out)
    )
```

```python
seq_len = 30
horizon = 10

X_past, X_future, y = make_dataset(data, seq_len, horizon)
```
